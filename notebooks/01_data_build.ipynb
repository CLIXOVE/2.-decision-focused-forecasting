# 01 Data Build

We generate a clean daily demand dataset so the project focuses on decision logic.

The goal is not domain realism.
The goal is to show how decisions can be evaluated and improved with forecasts.


```python
import numpy as np
import pandas as pd
from pathlib import Path

PROC = Path("../data/processed")
PROC.mkdir(parents=True, exist_ok=True)

np.random.seed(42)

dates = pd.date_range("2021-01-01", periods=730, freq="D")
trend = np.linspace(100, 140, len(dates))
season = 15 * np.sin(2 * np.pi * dates.dayofyear / 365)
noise = np.random.normal(0, 8, len(dates))

demand = trend + season + noise

df = pd.DataFrame({"date": dates, "demand": demand})
df.to_parquet(PROC / "daily_demand.parquet", index=False)

df.head()
```
